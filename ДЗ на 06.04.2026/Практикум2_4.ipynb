```python
import math
import numpy as np
import matplotlib.pyplot as plt

class R0:
    @staticmethod
    def distation(x,y):
        if len(x) != len(y):
            raise ValueError("|| Objects must have same dimensions")
        if len(x) == 0:
            raise ValueError("|| Objects must have at least 1 dimention")
        return max(abs(x[k]-y[k]) for k in range(len(x)))
    @staticmethod
    def distation_np(x,y):
        if len(x) != len(y):
            raise ValueError("|| Objects must have same dimensions")
        if len(x) == 0:
            raise ValueError("|| Objects must have at least 1 dimention")
        return np.max(np.abs(np.array(x)-np.array(y)))
class R1:
    @staticmethod
    def distation(x,y):
        if len(x) != len(y):
            raise ValueError("|| Objects must have same dimensions")
        if len(x) == 0:
            raise ValueError("|| Objects must have at least 1 dimention")
        return sum(abs(x[k]-y[k]) for k in range(len(x)))
    @staticmethod
    def distation_np(x,y):
        if len(x) != len(y):
            raise ValueError("|| Objects must have same dimensions")
        if len(x) == 0:
            raise ValueError("|| Objects must have at least 1 dimention")
        return np.sum(np.abs(np.array(x)-np.array(y)))
class R2:
    @staticmethod
    def distation(x,y):
        if len(x) != len(y):
            raise ValueError("|| Objects must have same dimensions")
        if len(x) == 0:
            raise ValueError("|| Objects must have at least 1 dimention")
        return math.sqrt(sum((x[k]-y[k])**2 for k in range(len(x))))
    @staticmethod
    def distation_np(x,y):
        if len(x) != len(y):
            raise ValueError("|| Objects must have same dimensions")
        if len(x) == 0:
            raise ValueError("|| Objects must have at least 1 dimention")
        return np.sqrt(np.sum((np.array(x)-np.array(y))**2))
        
def plot_R_2_shar_closed(metric,ax,x0,r,color):
    if len(x0) != 2:
        raise ValueError("|| Central object must have 2 dimentions")
    x_np = np.arange((x0[0]-r),(x0[0]+r)+.01,.01)
    y_np = np.arange((x0[1]-r),(x0[1]+r)+.01,.01)
    dots = []
    if   metric == 'R0':
        for x in x_np:
            for y in y_np:
                if R0.distation_np((x,y),x0) <= r:
                    dots.append((x,y))
    elif metric == 'R1':
        for x in x_np:
            for y in y_np:
                if R1.distation_np((x,y),x0) <= r:
                    dots.append((x,y))
    elif metric == 'R2':
        for x in x_np:
            for y in y_np:
                if R2.distation_np((x,y),x0) <= r:
                    dots.append((x,y))
    else:
        raise ValueError(f"|| '{metric}' metric does not supported, only: 'R0', 'R1' and 'R2'")
    ax.plot([d[0] for d in dots],[d[1] for d in dots],color=color)
```

```python
import math

def solve_fx_is_x(f,alpha,x1,epsi,max_iter):
    x_n = [x1]
    for k in range(1,max_iter+1):
        x_n.append(f(x_n[-1]))
        diff  = abs(x_n[-1]-x_n[-2])
        error = (alpha*diff)/(1-alpha)
        if k <= 7 or error < epsi or k == max_iter:
            print(f"{k} iteration: x ≈ {x_n[-1]:<8.7} with difference {diff:<9.7} & error {error:<9.7}")
        if error < epsi:
            print(f"Done! Deep (k) is {k} iterations & x ≈ {x_n[-1]}")
            return (x_n,k)
    print("Nothing found...")
    return (x_n,max_iter)

f      = lambda x: math.atan(x) + math.pi
x_n, k = solve_fx_is_x(f,0.2884,3,0.0001,1000)
```
