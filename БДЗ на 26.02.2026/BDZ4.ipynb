**Задание 1**

```python
import numpy as np
import matplotlib.pyplot as plt

y1 = lambda x:  x**2 + 4*x + 1
y2 = lambda x: -x**2 + 7*x + 3

x  = np.arange(-7,3,.01)
plt.plot(x,y1(x),color='seagreen',label='$x^2 + 4x + 1$')
x  = np.arange(-1,8,.01)
plt.plot(x,y2(x),color='DarkOrchid',label='$-x^2 + 7x + 3$')
x  = np.arange(-1/2,2,.01)
plt.fill_between(x, y1(x), y2(x), color='DarkSlateBlue', alpha=.4)
plt.grid(); plt.legend()

import sympy as sp

x  = sp.Symbol('x')
y  = y2(x) - y1(x)
i  = sp.integrate(y,x)
sp.pprint(i)
print("Final answer:",end='\t'); sp.pprint(sp.integrate(y,(x,-1/2,2)))
```

**Задание 2**

```python
import numpy as np
import matplotlib.pyplot as plt

q  = np.arange(0,2*np.pi+.001,.001)
r1 = 4*np.cos(3*q)
r2 = [2]*len(q)

plt.polar(q, r1, color = 'DarkOrchid', label='$r = 4\cos{3\phi}$')
plt.polar(q, r2, color = 'DarkSlateBlue',   label='$r = 2$')
plt.fill_between(q, r1, r2, where=(r2 <= r1), color='CadetBlue', alpha=.5)
plt.gca().set_rlim(0, None)
plt.legend()

import sympy as sp

q = sp.Symbol('q')
i = sp.integrate(((4*sp.cos(3*q))**2-4),(q,0,sp.pi/9))/2
ans = 3*i
ans_str = sp.pretty(ans).split('\n')
print("Final answer:   ",ans_str[0])
print(*(' '*17+s for s in ans_str[1:]),sep='\n')
```

**Задание 3**

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(np.sqrt(3),np.sqrt(8),.01)
y = np.log(5/(2*x))
plt.plot(x,y,color='DarkOrchid',label=r'$\ln{\frac{5}{2x}}$')
plt.axis('equal'); plt.grid(); plt.legend()

import sympy as sp

x = sp.Symbol('x')
i = sp.integrate(sp.sqrt(x**2+1)/x,x)
sp.pprint(i)
dl = sp.integrate(sp.sqrt(x**2+1)/x,(x,np.sqrt(3),np.sqrt(8)))
ans_str = sp.pretty(dl).split('\n')
print("Final answer:   ",ans_str[0])
print(*(' '*17+s for s in ans_str[1:]),sep='\n')
```

**Задание 6**

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

x = np.linspace(0,1,400)
y = np.exp(1-x)
plt.plot(x,y,label='e^{1-x}')
plt.axhline(0,color='k'); plt.grid(); plt.legend()

x_sym = sp.symbols('x', real=True)
y_sym = sp.exp(1-x_sym)
V = sp.pi*sp.integrate(y_sym**2, (x_sym, 0, 1))
sp.pprint(sp.simplify(V))
```

**Задание 7**

```python
import sympy as sp

x, a = sp.Symbol('x'), sp.Symbol('a',positive=True)
y = sp.sqrt((a+x)/(a-x))
i = sp.integrate(y,(x,0,a))
ans_str = sp.pretty(i).split('\n')
print("Final answer:   ",ans_str[0])
print(*(' '*17+s for s in ans_str[1:]),sep='\n')
```
