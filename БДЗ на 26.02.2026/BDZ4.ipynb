**Задание 1**

```python
import numpy as np
import matplotlib.pyplot as plt

y1 = lambda x:  x**2 + 4*x + 1
y2 = lambda x: -x**2 + 7*x + 3

x  = np.arange(-7,3,.01)
plt.plot(x,y1(x),color='seagreen',label='$x^2 + 4x + 1$')
x  = np.arange(-1,8,.01)
plt.plot(x,y2(x),color='DarkOrchid',label='$-x^2 + 7x + 3$')
x  = np.arange(-1/2,2,.01)
plt.fill_between(x, y1(x), y2(x), color='DarkSlateBlue', alpha=.4)
plt.grid(); plt.legend()

import sympy as sp

x  = sp.Symbol('x')
y  = y2(x) - y1(x)
i  = sp.integrate(y,x)
sp.pprint(i)
print("Final answer:",end='\t'); sp.pprint(sp.integrate(y,(x,-1/2,2)))
```

**Задание 2**

```python
import numpy as np
import matplotlib.pyplot as plt

q  = np.arange(0,2*np.pi+.001,.001)
r1 = 4*np.cos(3*q)
r2 = [2]*len(q)

plt.polar(q, r1, color = 'DarkOrchid', label='$r = 4\cos{3\phi}$')
plt.polar(q, r2, color = 'DarkSlateBlue',   label='$r = 2$')
plt.fill_between(q, r1, r2, where=(r2 <= r1), color='CadetBlue', alpha=.5)
plt.legend()
plt.gca().set_rlim(0, None)

import sympy as sp

q = sp.Symbol('q')
i = sp.integrate(((4*sp.cos(3*q))**2-4),(q,0,sp.pi/9))/2
ans = 3*i
ans_str = sp.pretty(ans).split('\n')
print("Final answer:   ",ans_str[0])
print(*(' '*17+s for s in ans_str[1:]),sep='\n')
```

**Задание 3**

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(np.sqrt(3),np.sqrt(8),.01)
y = np.log(5/(2*x))
plt.plot(x,y,color='DarkOrchid',label=r'$\ln{\frac{5}{2x}}$')
plt.axis('equal'); plt.grid(); plt.legend()

import sympy as sp

x = sp.Symbol('x')
i = sp.integrate(sp.sqrt(x**2+1)/x,x)
sp.pprint(i)
dl = sp.integrate(sp.sqrt(x**2+1)/x,(x,np.sqrt(3),np.sqrt(8)))
ans_str = sp.pretty(dl).split('\n')
print("Final answer:   ",ans_str[0])
print(*(' '*17+s for s in ans_str[1:]),sep='\n')
```

**Задание 4**

```python
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

t = np.arange(0, .51, .01)
x = t
y = np.arcsin(t)
z = .25*np.log((1+t)/(1-t))
plt.plot(x, y, z, color='DarkOrchid', label=r'$x = t$  $y = \arcsin{t}$  $z = \frac{1}{4}\ln{\frac{1+t}{1-t}}$')
plt.grid(); plt.legend()

import sympy as sp

t = sp.Symbol('t')
x = t
y = sp.asin(t)
z = sp.Rational(1,4)*sp.log((1+t)/(1-t))

x_p = sp.diff(x,t)
y_p = sp.diff(y,t)
z_p = sp.simplify(sp.diff(z,t))

integr_f = sp.sqrt(x_p**2 + y_p**2 + z_p**2)
L = sp.integrate(integr_f, (t,0,sp.Rational(1,2)))
sp.pprint(L)
```

**Задание 5**

```python
import numpy as np
import matplotlib.pyplot as plt

q = np.linspace(0, np.pi/2, 400)
r = np.cos(q) + np.sin(q)
plt.polar(q, r, color='DarkOrchid',label=r'$r = \cos{\phi} + \sin{\phi}$')
plt.legend()

import sympy as sp

q = sp.symbols('q')
r = sp.cos(q) + sp.sin(q)
r_p = sp.diff(r, q)
integr_f = sp.sqrt(r**2 + r_p**2)
L = sp.integrate(integr_f, (q, 0, sp.pi/2))
sp.pprint(sp.simplify(L))
```

```python
import numpy as np
import matplotlib.pyplot as plt

## Сечение

fig1, ax1 = plt.subplots(figsize=(5,5))

x = np.linspace(0, 1, 800)
ax1.plot(x, np.exp(1-x), color='DarkOrchid', label='$y = e^{1-x}$')
ax1.axvline(0, color='DarkSeaGreen', label='$x = 0$')
ax1.axvline(1, color='DarkSeaGreen', label='$x = 1$')
ax1.axhline(0, color='DarkSlateBlue', linestyle='--', label='$y = 0$')

ax1.fill_between(x, 1, np.exp(1-x), color='CadetBlue', alpha=.56)

ax1.set_xlabel('X'); ax1.set_ylabel('Y')
ax1.set_title('Сечение в плоскости OXY')
plt.axis('equal'); plt.grid(); plt.legend()
plt.tight_layout()

## Тело вращения

fig2 = plt.figure(figsize=(10,8))
ax2  = fig2.add_subplot(111, projection='3d')

u = np.linspace(0, 1, 800)
v = np.linspace(0, 2*np.pi, 800)
u, v = np.meshgrid(u, v)

r = np.exp(1-u)
x = u
y = r * np.cos(v)
z = r * np.sin(v)
ax2.plot_surface(x, y, z, color='DarkOrchid', alpha=.7)

ax2.set_xlabel('X'); ax2.set_ylabel('Y'); ax2.set_zlabel('Z')
ax2.set_title('Тело вращения вокруг оси OX')
plt.tight_layout()

import sympy as sp

x = sp.Symbol('x')
y = sp.exp(1-x)
V = sp.pi*sp.integrate(y**2, (x, 0, 1))
sp.pprint(sp.simplify(V))
```
